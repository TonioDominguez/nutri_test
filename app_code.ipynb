```python
import pandas as pd
```

```python
data = pd.read_csv("recetas_ingredientes_cantidades.csv")
```

```python
df = data.copy()
```

```python
# df.cantidad se importa en float, convertimos a integer

df["cantidad"] =df["cantidad"].astype("Int64")
```

```python
df
```

### IMPRESIÓN DE MENÚ

```python
#Listado de menú disponible

for x in df["nombre"].unique():
    print(f"\n- {x}")
```

```python
# Visualiza receta

def mostrar_receta(df, receta):
    
    receta_filtrada = df[df["nombre"].str.contains(receta, case=False)]
    
    if len(receta_filtrada) > 0:
        print(f"\nIngredientes para {receta_filtrada['nombre'].iloc[0]}:")
        print ("-" * 50)
        
        for idx, row in receta_filtrada.iterrows():
            if pd.isna(row['cantidad']) and pd.isna(row['unidad']):
                print(f"- {row['ingrediente']}")
            else:
                print(f"- {row['ingrediente']} {row['cantidad']} {row['unidad']}")
    else:
        print("No existe esta receta")
```

```python
while True:
    receta = input("¿Cuál es la receta?: ")
    if receta.lower() == "salir":
        break
        
    mostrar_receta(df, receta)
    print("\nEscribe 'salir' para cerrar el programa")
```
